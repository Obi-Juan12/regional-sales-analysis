```python
# The first row of df_state_reg contains the real column names.
new_header = state_regions_df.iloc[0]      # grab the first row
state_regions_df.columns = new_header             # set columns
state_regions_df  = state_regions_df[1:].reset_index(drop=True)  # drop header row and reset
```

```python
# Data Cleaning


# Merge with Customers

main_df = pd.merge(
    sales_orders_df,
    customers_df,
    how="left",
    left_on="Customer Name Index",   # key in sales_orders
    right_on="Customer Index"        # key in customers
)



# Merge with Products

main_df = pd.merge(
    main_df,
    products_df,
    how="left",
    left_on="Product Description Index",
    right_on="Index"
)




# Merge with Regions

main_df = pd.merge(
    main_df,
    regions_df,
    how="left",
    left_on="Delivery Region Index",
    right_on="id"
)



# Merge with State Regions

main_df = pd.merge(
    main_df,
    state_regions_df,
    how="left",
    left_on="state_code",
    right_on="State Code"
)


# Merge 2017 Budgets

main_df = main_df.merge(
    budgets_df,
    how='left',
    on='Product Name'
)


main_df.head()
```

```python
#Creating Metrics

# Convert OrderDate to datetime
main_df["OrderDate"] = pd.to_datetime(main_df["OrderDate"], errors="coerce")

# Create time-based fields
main_df["Year"] = main_df["OrderDate"].dt.year
main_df["Month"] = main_df["OrderDate"].dt.month
main_df["YearMonth"] = main_df["OrderDate"].dt.to_period("M")

# Financial metrics
main_df["Revenue"] = main_df["Line Total"]
main_df["Cost"] = main_df["Order Quantity"] * main_df["Total Unit Cost"]
main_df["Profit"] = main_df["Revenue"] - main_df["Cost"]

# Calculate Profit Margin %
main_df["Margin %"] = np.where(main_df["Revenue"] > 0, main_df["Profit"] / main_df["Revenue"], 0)
```

```python
# Drop multiple columns
main_df = main_df.drop(columns=["Currency Code", "Warehouse Code", "Delivery Region Index", "Product Description Index", "land_area", "water_area", "Customer Name Index", "name", "county", "state", "population", "households"])
```

```python
#Checking data quality once again to verify data
print(main_df.shape)          # rows  columns
print(main_df.columns)        # what columns we now have
print(main_df.isna().sum())   # missing values
print(main_df.head())         # preview
```

```python
#Exploratory Analysis

# Revenue share by channel (Pie)
kpi_channel = (main_df.groupby("Channel")
                 .agg(Revenue=("Revenue","sum"))
                 .sort_values("Revenue", ascending=False)
                 .reset_index())

plt.figure(figsize=(6,6))
plt.pie(kpi_channel["Revenue"].values, autopct="%1.1f%%", startangle=90)
plt.title("Revenue Share by Channel")
plt.legend(kpi_channel["Channel"].astype(str).tolist(), loc="best", bbox_to_anchor=(1.1, 0.5))
plt.tight_layout()
plt.show()

print(kpi_channel)
```

# **Key Insights:**


*   The pie chart shows that Wholesale is the biggest contributor with revenue of $668 million.
*   Second biggest contributor is Distributor with $387 million in revenue.
* Exports can provide growth opportunities in international markets this can diversify the company and reduce risk.


```python
# Top 10 products by revenue (Horizontal bar)

kpi_product = (main_df.groupby("Product Name")
                 .agg(Revenue=("Revenue","sum"), Profit=("Profit","sum"))
                 .sort_values("Revenue", ascending=False)
                 .reset_index())

top10 = kpi_product.head(10).iloc[::-1]   # largest on top

plt.figure(figsize=(9,4))

# Normalize revenues to map them onto a colormap
norm = plt.Normalize(top10["Revenue"].min(), top10["Revenue"].max())
colors = cm.viridis(norm(top10["Revenue"]))   # sequential colors based on value

plt.barh(top10["Product Name"].astype(str),
         top10["Revenue"],
         color=colors)

plt.title("Top 10 Products by Revenue (Millions)")
plt.xlabel("Revenue")

# Format x-axis labels as millions with whole numbers
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x/1e6:.0f}'))


plt.tight_layout()
plt.show()

print(kpi_product)
```

# **Key Insights:**


*   Product 26 and 25 are strong leaders with revenue for Product 26 being $117 million and for product 25 being $109 million. This makes up 18% of total company revenue.
*   If any demand of the top 5 products fall it can significantly impact revenue.
* Product 1-4 grouped together are showing revenue but looking at individually they do not rival the top 2 products.



```python
# Total Revenue by Region

# Aggregate by region
region_perf = main_df.groupby("Region").agg(
    Revenue=("Revenue", "sum"),
    Profit=("Profit", "sum"),
    Margin=("Margin %", "mean")
).reset_index()

# Plot horizontal bar
plt.figure(figsize=(9,4))
plt.barh(region_perf["Region"], region_perf["Revenue"], color="skyblue")
plt.title("Total Revenue by Region (Millions)")
plt.xlabel("Revenue")
# Format x-axis labels as millions with whole numbers
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x/1e6:.0f}'))
plt.tight_layout()
plt.show()

# Profit plot
plt.figure(figsize=(9,4))
plt.barh(region_perf["Region"], region_perf["Profit"], color="lightgreen")
plt.title("Total Profit by Region (Millions)")
plt.xlabel("Profit")
# Format x-axis labels as millions with whole numbers
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x/1e6:.0f}'))
plt.tight_layout()
plt.show()
```

# **Key Insights:**


* Based on both of the charts the West region leasds in both revenue $372 million  and profit of $139 million.
  
* The South and Midwest regions demonstrates in being strong contributors.
* The Northeast region demonstrates weak contribution and can be a region to focus on to increase business profit and revenue.

```python
# Top 10 States by Number of Orders
state_orders = main_df.groupby("State").agg(
    Orders=("OrderNumber", "nunique")
).sort_values("Orders", ascending=False).head(10)

# Create sequential green shades using colormap
colors = plt.cm.Greens(np.linspace(0.4, 0.9, len(state_orders)))

# Vertical Bar Chart with Sequential Green Colors
plt.figure(figsize=(10,6))
plt.bar(state_orders.index, state_orders["Orders"], color=colors)
plt.title("Top 10 States by Number of Orders")
plt.xlabel("State")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Display the values
print(state_orders)
```

# Key Insights:


*   California has the highest number of orders with orders of 7,613, California is critical to maintain service levels and competitive positioning
*   States like Illionis, Florida, and Texas are also strong markets



```python
# Exclude year 2018 ---
df_no2018 = main_df[main_df["Year"] != 2018].copy()

# Aggregate total revenue by calendar month (sum across years, excl. 2018) ---
monthly_profile_total = (
    df_no2018.groupby("Month")
             .agg(Revenue=("Revenue", "sum"))
             .reset_index()
             .sort_values("Month")
)

# Convert revenue to millions
monthly_profile_total["Revenue_Millions"] = monthly_profile_total["Revenue"] / 1e6

# Add month names for clarity
monthly_profile_total["MonthName"] = pd.to_datetime(
    monthly_profile_total["Month"], format="%m"
).dt.strftime("%b")

# Plot seasonality profile ---
plt.figure(figsize=(13,4))
plt.plot(
    monthly_profile_total["MonthName"],
    monthly_profile_total["Revenue_Millions"],
    marker="o",
    linewidth=2,
    color="darkred"
)
plt.title("Total Monthly Sales Trend (Excluding 2018)")
plt.xlabel("Month")
plt.ylabel("Total Revenue (Millions)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()

# Display values ---
print(monthly_profile_total)

```

# **Key Insights:**


* The strongest month is May with $102.3 million

* The weakest month is February with $91.3 million, which shows a demand dip and can be caused by post-holiday slowdowns.
* The months of November and December remain strong, which marketing, promotions, and inventory stocking should be prioritized in these months.
* For weak periods like in February and April, special promotions or targets campaigns to increase demand.
